```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# carregando a base de dados
df = pd.read_csv("imoveis.csv", sep=",", encoding="utf-8")

correlacoes = df.corr()
fig, ax = plt.subplots()
ax.set_title('Correlação entre as variáveis')
sns.heatmap(correlacoes,square=True, annot=True)
plt.show()
```
